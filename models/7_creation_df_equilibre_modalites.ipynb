```python
import pandas as pd
df = pd.read_csv('../data/avis/general_df_clean.csv')
```

```python
df.head()
```

```python
mod_etoiles = df.etoiles.value_counts().sort_values()
n_mod = mod_etoiles.min()
```

```python
# mélanger le jeu de données
```

```python
# Mélange aléatoire des lignes de la df
df = df.sample(frac = 1, random_state=7)
```

```python
df1 = df[df.etoiles ==1].iloc[0:n_mod,:]
df2 = df[df.etoiles ==2].iloc[0:n_mod,:]
df3 = df[df.etoiles ==3].iloc[0:n_mod,:]
df4 = df[df.etoiles ==4].iloc[0:n_mod,:]
df5 = df[df.etoiles ==5].iloc[0:n_mod,:]

len(df2)
```

```python
df_clean_equal = pd.concat([df1,df2,df3,df4,df5])
```

```python
# On mélange à nouveau 
df_clean_equal = df_clean_equal.sample(frac = 1, random_state=7)
df_clean_equal.head()
```

```python
df_clean_equal.shape
```

```python
df_clean_equal.etoiles.value_counts()
```

```python
# Well balanced as all things should be
# sauvegarde

df_clean_equal.to_csv("../data/avis/general_df_clean_equil.csv")
```
