```python
# Refined thematic codebook (labels)
codebook2 = pd.read_excel('../data/avis/general_df_clean_sent_15k_manual_code.xlsx', sheet_name = "codebook2", index_col=0, header=0)
codebook2.style.hide()
```

```python
# same for the validation dataset
test = pd.read_excel('../data/avis/general_df_clean_sent_15k_manual_code.xlsx', sheet_name="test", index_col=0, header = 0)
test['codes agrégés'] = test['codes agrégés'].fillna("none")
test['codes agrégés'] =[codes.replace("na","none") for codes in test['codes agrégés']]
test['codes agrégés'] =[codes.replace(" ","") for codes in test['codes agrégés']]
codebook2['MERGED_CODES'] = [codes.replace(" ","") for codes in codebook2['MERGED_CODES']]

#creating new columns for the new labels
test[[codebook2['CODE'].tolist()]] = 9

# looping over codebook2 and labeled dataframe to dichotomize labels in the new coding (labeling) tree
for codebook_i, ref_codes in enumerate(codebook2['MERGED_CODES']):
    ref_codes = ref_codes.split(",")
    var_name = str(codebook2['CODE'].iloc[codebook_i])
    print(f'----- For label {var_name} using reference labels {ref_codes}')
    for review_i, labeled_codes in enumerate(test['codes agrégés']):
        labeled_codes = labeled_codes.split(";")
        resonate = any(item in labeled_codes for item in ref_codes)
        if resonate:
            print(f"🔥 match between labels {labeled_codes} and {ref_codes} (reference labels)")
            test[var_name].iloc[review_i] = 1
        else:
            print(f"no match between labels {labeled_codes} and {ref_codes} (reference labels)")
            test[var_name].iloc[review_i] = 0

test.head()

with pd.ExcelWriter('/Users/lheyerda/Library/CloudStorage/OneDrive-InstitutPasteurParis/github/juil23_cds_supply_chain/data/avis/general_df_clean_sent_15k_manual_code.xlsx', engine='openpyxl', mode='a') as writer:
    # Write the new DataFrame to a new sheet
    test.to_excel(writer, sheet_name='extra_test2', index=False)
```
